```python
import numpy as np

X = np.array([[1, 3], [2, 2], [3, 1], [4, 4], [5, 5], [6, 6]])
y = np.array([0, 0, 0, 1, 1, 1])

m0 = X[y == 0].mean(axis=0)
m1 = X[y == 1].mean(axis=0)

cx1 = X[y == 0] - m0
cx2 = X[y == 1] - m1
S1 = np.dot(cx1.T, cx1) / (X[y == 0].shape[0] - 1)
S2 = np.dot(cx2.T, cx2) / (X[y == 1].shape[0] - 1)

S_w = S1 + S2

mean_diff = m1 - m0

S_w_inv = np.linalg.inv(S_w)

w_roof = np.dot(S_w_inv, mean_diff)

print("Discriminant projection:", w_roof)
```
